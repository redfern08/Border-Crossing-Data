# Border Crossing Data 2020-2023

The latest reports of border crossings between the US-Mexico and US-Canadian borders.

```python
import pandas as pd
import requests as r
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from datetime import datetime, date, time
```

```python
df = pd.read_csv("data/Border_Crossing_Entry_Data.csv")
```

```python
pd.set_option('display.max_rows', 61)
```

First, I wanted to gather the data via a CSV file linked from Data.gov through the Department of Transportation dataset showing Border crossings.  The data set goes back to 1997 but for this purpose, I wanted to show the border crossings that happened from January 2020 to August 2023.

```python
df = df.head(61)
```

```python
df
```

After we establish the timeline for the data we want, I will then go ahead and only gather the necessary columns and rows.  I want the state, border, date, measure (method of transport), and value (how many have crossed).  I have also collected the latitude and longitude so I can use the coordinates to display an outline of where the crossings have occurred.

```python
df = pd.DataFrame(df[['State', 'Border', 'Date', 'Measure', 'Value', 'Latitude', 'Longitude']])
df
```

Next I need to group the states and measures together and create a sum total of crossings using the values provided.

```python
grouped = df.groupby(['State', 'Border', 'Measure'])
grouped.sum()
```

Here I created variables to use for the visualization.  Typically any variables I create would have been stored early on however, I was not sure what I wanted to visualize so for me, it was best to wait until I had a better understanding of what I wanted to display.

```python
# Create variables to use for later.  All we want is state, border and measure.  Later we will work will both the value and date of each crossing.

state = df['State']
measure = df['Measure']
total = df['Value']
border = df['Border']
lon = df['Longitude']
lat = df['Latitude']
grouped_data = grouped.first()
```

```python
# plot the data using matplotlib.  the c=variable and s=variable will use the total variables to display the points of the total values of crossings.
plt.rcParams['figure.figsize']=(20,10)

plt.style.use('ggplot')

x = lon
y = lat

fig, ax = plt.subplots()

ax.scatter(x, y, c=total, s=total * 0.05, cmap='inferno', alpha=.725)
plt.title('Border Crossing Data of US-Mexico and US-Canada Borders in 2020-2023')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.margins(.15, .15)


plt.show()
```
