# Border Crossing Data 2020-2023

The latest reports of border crossings between the US-Mexico and US-Canadian borders.

```python
import pandas as pd
import requests as r
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
pd.set_option('display.max_rows', 61)
```

```python
df = pd.read_csv("data/Border_Crossing_Entry_Data.csv")
```

First, I wanted to gather the data via a CSV file linked from Data.gov through the Department of Transportation dataset showing Border crossings.  The data set goes back to 1997 but for this purpose, I wanted to show the border crossings that happened from December 2020 to August 2023.

```python
df = df.head(61)
```

```python
df
```

```python
# Create variables to use for later.  All we want is state, border and measure.  Later we will work will both the value and date of each crossing.

state = df['State']
measure = df['Measure']
total = df['Value']
date = df['Date']
lon = df['Longitude']
lat = df['Latitude']
```

```python
plt.rcParams['figure.figsize']=(30,20)

plt.style.use('ggplot')

x = lon
y = lat

fig, ax = plt.subplots()

ax.scatter(x, y, c=total, s=total)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```
